```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

csv_file = '/content/drive/Shared drives/ABCD_Model/Explainability/no_demos_feature_ablation_results.csv'
df_results = pd.read_csv(csv_file)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort by AUC
df_sorted = df_results.sort_values(by='AUC', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_sorted, x='ablated_feature', y='AUC', hue='feature_type', dodge=False, palette='muted')
plt.xticks(rotation=45, ha='right')
plt.title('No Demos AUC After Feature Ablation')
plt.ylabel('AUC')
plt.xlabel('Ablated Feature')
plt.ylim(bottom=0.6)
plt.tight_layout()
plt.show()
```

```python
metrics_to_plot = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC', 'NPV', 'Specificity']

df_melted = df_results.melt(id_vars=['ablated_feature', 'feature_type'], value_vars=metrics_to_plot,
                            var_name='Metric', value_name='Score')

plt.figure(figsize=(14, 7))
sns.lineplot(data=df_melted, x='ablated_feature', y='Score', hue='Metric', style='feature_type', markers=True)
plt.xticks(rotation=45, ha='right')
plt.title('No Demos Performance Metrics by Ablated Feature')
plt.tight_layout()
plt.show()
```

```python
pivot_df = df_results.set_index('ablated_feature')[metrics_to_plot]

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('no Demos Metric Values After Feature Ablation')
plt.ylabel('Ablated Feature')
plt.tight_layout()
plt.show()
```
